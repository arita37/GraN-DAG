# Showcase to show that the bnlearn impute function touches non NA values

My understanding of the bnlearn [impute](http://www.bnlearn.com/documentation/man/impute.html) function is that it will impute values that are missing, but will not touch values that are present in its `data` input data.frame. The below code shows that this understanding does not match reality.

Either my understanding of the purpose of the impute function is wrong or this is a bug in the impute function.
